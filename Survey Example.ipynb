```python
under20 = df[df['Age'] < 20]
over20 = df[(df['Age'] >= 20) & (df['Age'] < 30)]
over30 = df[(df['Age'] >= 30) & (df['Age'] < 40)]
over40 = df[(df['Age'] >= 40) & (df['Age'] < 50)]
over50 = df[df['Age'] > 50]
print(under20['Trans'].value_counts(normalize=True))
print(over20['Trans'].value_counts(normalize=True))
print(over30['Trans'].value_counts(normalize=True))
print(over40['Trans'].value_counts(normalize=True))
print(over50['Trans'].value_counts(normalize=True))
```
